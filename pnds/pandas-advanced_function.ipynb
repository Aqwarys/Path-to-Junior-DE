```python
import pandas as pd
```

```python
df = pd.read_excel("data/crm.xlsx")
```

```python
# подготовка данных
crm_stats = df.groupby("Покупка")[["Сумма", "Возраст"]].agg(avg_sum=("Сумма", "mean"), avg_age=("Возраст", "mean")).reset_index()

crm_stats.to_excel("data/crm_stats.xlsx", index=True, startrow=1, index_label="id")
```

```python
# Работает только с одной колонкой (Series)
# Применяет функцию к каждому элементу этой колонки
df["Телефон"] = df["Телефон"].map(str)
# или даже словарь для замены значений
df["rating"] = df["rating"].map({"дг": "ДГ", "PG": "Teens", "R": "Adults"})
```

```python
df["Дата рождения"] = pd.to_datetime(df["Дата рождения"], errors="coerce")
df.dtypes
```

```python

# Работает с DataFrame или Series.
# Если применяешь к Series — то аналог map().
# Если применяешь к DataFrame — можешь указать axis=1, чтобы применить к каждой строке.
# Пример 1 — применить к одной колонке:
from datetime import date
today = date.today()
df["Город"] = df["Город"].apply(str.upper)

df["Возраст"] = df["Дата рождения"].apply(lambda x:
    today.year - x.year -
    ((today.month, today.day) < (x.month, x.day))
).astype("Int64")
df
```

```python
df["Карта клиента"] = df.apply(
    lambda row: f"Имя: {row["Имя"]}, Возраст: {row["Возраст"] if row["Возраст"] is not None else "Неизвестно"}", axis=1
)
```
