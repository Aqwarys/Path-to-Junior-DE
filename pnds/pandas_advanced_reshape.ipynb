```python
import pandas as pd
```

```python
df = pd.read_csv("data/updated_anime.csv", encoding="utf-8")
df.head(2)
```

```python

# df.pivot(
#     index="...",     # Что станет строками
#     columns="...",   # Что станет столбцами
#     values="..."     # Что положить в ячейки
# )


# ✔ Важно:
# pivot НЕ умеет агрегировать
# Если в одном (year, type) несколько строк → будет ошибка
df.pivot(
    index="year",
    columns="type",
    values="score"
)
```

```python
# df.pivot_table(
#     index="...",
#     columns="...",
#     values="...",
#     aggfunc="mean"   # или sum, count, max, min …
# )


# ✔ pivot_table:
# НЕ падает при дубликатах
# сам считает mean/sum
# всегда безопасный вариант
df.pivot_table(
    index="year",
    columns="type",
    values="score",
    aggfunc="mean"
)
```

```python
# Multiindex
df.groupby(["year", "type"]).agg({"score": "mean"})
```
