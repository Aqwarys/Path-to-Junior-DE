```python
import pandas as pd
```

```python
df = pd.read_csv("data/anime.csv", encoding="utf-8")
df.head(3)
# & = И
# | = ИЛИ
# ~ = НЕ
# (вокруг каждого условия нужны скобки!)
```

```python
# Series
df["title"]

# Dataframe
df[["title", "score", "members"]]

# Add new column
df["popularity"] = df["score"] * (df["members"] / 10000)
df[["title", "popularity"]]
```

```python
# Filtering by condition
high_score = df[df["score"] > 9]
# Dataframe with columns
high_score
```

```python
# Комбинированные фильтры
tv_high_score = df[(df["type"] == "TV") & (df["score"] > 9)]
tv_high_score
```

```python
not_tv = df[df["type"] != "TV"]
not_tv
```

```python
# Проверка на вхождение
main_types = df[df["type"].isin(["TV", "Movie"])]
# Отрицательное
not_main_types = df[~df["type"].isin(["TV", "Movie"])]
```

```python
# Пропуски DataFrame
no_score = df[df["score"].isna()]
```

```python
# loc — работает по меткам,
# iloc — по позициям (номерам строк и колонок).

# По имени индекса и столбца
df.loc[0:5, ["title", "score"]]

# # По числовому индексу
df.iloc[0:5, [1, 3, 6]]

# По условию + выбор колонок
df.loc[df["score"] > 8, ["title", "score", "members"]]
```

```python
# Sorting single column
df.sort_values("score", ascending=False).head(5)
# Sorting multiple columns
df.sort_values(["score", "members"], ascending=[False, True]).head(5)
```

```python
# Check existing index
df.index

# Set index to new one
df.set_index("mal_id", inplace=True)
# Resetting index and place new range index
df.reset_index(inplace=True)
```
