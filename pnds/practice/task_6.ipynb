```python
import pandas as pd
```

```python
df = pd.read_csv("../data/anime.csv", encoding="utf-8")
```

```python
rating_group = {
    "PG-13 - Teens 13 or older": "PG-13",
    "PG - Children" : "PG",
    "R - 17+ (violence & profanity)" : "R",
    "R+ - Mild Nudity" : "R+",
    "G - All Ages" : "G"
}

df["rating"] = df["rating"].map(rating_group).fillna("unknown")
```

```python
df["year"] = df["aired"].str.extract(r"(\d{4})").astype("Int64")
df
```

```python
def age_group(year):
    if year >= 2018:
        return "new"
    elif year >= 2010:
        return "mid"
    elif year < 2010:
        return "old"
    return "unkown"


df["age_group"] = df["year"].apply(age_group)
```

```python
df["score_norm"] = df["score"] - df.groupby("type")["score"].transform("mean")
df
```

```python
import re

def parse_duration(s):
    if not s:
        return None

    hours = re.search(r"(\d+)\s*hr", s)
    minutes = re.search(r"(\d+)\s*min", s)

    h = int(hours.group(1)) if hours else 0
    m = int(minutes.group(1)) if minutes else 0

    return h * 60 + m

df["duration_minutes"] = df["duration"].apply(parse_duration)
```

```python
df["hype_score"] = df["popularity"] / df.groupby("type")["popularity"].transform("mean")
```

```python
overhyped = df[df["hype_score"] > 1]
overhyped
```
