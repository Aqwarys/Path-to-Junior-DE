```python
import pandas as pd
```

```python
df = pd.read_csv("../data/anime.csv", encoding="utf-8")
df.head(2)
```

```python
df_tv = df.loc[
    (df["type"] == "TV") &
    (df["score"] > 8) &
    (df["episodes"] >= 12)
]
df_tv.dtypes
```

```python
df_tv["year"] = df_tv["aired"].str.extract(r"(\d{4})").astype("Int64")
df_tv[["title", "type", "status", "rating", "year"]]
```

```python
df_tv_cleaned = df_tv.dropna(subset=["score", "rating"])
print(f"Before: {len(df_tv)}, after: {len(df_tv_cleaned)}")
```

```python
df_tv_cleaned["popularity"] = df_tv_cleaned["score"] * (df_tv_cleaned["members"]/10000)
df_tv_cleaned[["title", "type", "status", "rating", "year", "popularity"]]
```

```python
avg_scores = (
    df_tv_cleaned.groupby("type")["score"]
        .agg(["mean", "count"])
        .reset_index()
        .sort_values(by="mean", ascending=False)
)

avg_scores
```
