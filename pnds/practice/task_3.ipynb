```python
import pandas as pd
from pathlib import Path
```

```python
files = list(Path("../data/task3").rglob("*.csv"))
print(f'Найдено:\n{"\n".join(map(str, files))}')
try:
    df = pd.concat(map(pd.read_csv, files), ignore_index=True)
    df.to_csv("../data/task3/combined.csv", index=False)
    print(f"Данные объединены и сохранены в файл combined.csv\nShape: {df.shape}")

except Exception as e:
    print(f"Произошла ошибка: {e}")
```

```python
try:
    df.drop_duplicates(subset=["mal_id"], inplace=True)
    df.dropna(subset=["score"], inplace=True)
    df["score"], df["members"] = df["score"].astype("float"), df["members"].astype("Int64")
    df["year"] = df["aired"].str.extract(r"(\d{4})").astype("Int64")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```

```python
df_filtered = df.loc[
    (df["episodes"] > 0) &
    (df["score"] > 0) &
    (df["score"].max() <= 10) &
    (df["members"] > 0)
]
df_filtered = df[df["type"].isin(["TV", "Movie", "OVA", "ONA"])]
```

```python
df.groupby(["type", "year"]).agg(
    {"score": ["mean",]},
    {"members": ["sum",]},
    {"*": ["count",]}
).reset_index()
```
