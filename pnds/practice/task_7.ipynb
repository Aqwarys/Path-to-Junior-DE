```python
import pandas as pd
```

```python
df = pd.read_csv('../data/updated_anime.csv', encoding="utf-8")
```

```python
df.pivot_table(
    index="year",
    columns="type",
    values="score",
    aggfunc="mean"
).sort_values("year", ascending=False)
```

```python
df.pivot(
    index="year",
    columns="type",
    values="score"
)
# ValueError: Index contains duplicate entries, cannot reshape  жто потому что индекс или колонный повтарябтся
# Так как невозможно поставить два знаечения в одно поле без аггрегаций так что оно падает и не имеет смысла
# создавать дубликаты
```

```python
agg = df.groupby(["year", "type"]).agg(
    total_members=("members", "sum")
)
popular_type = agg.loc[agg.groupby("year")["total_members"].idxmax()].drop("total_members", axis=1)
popular_type
```
